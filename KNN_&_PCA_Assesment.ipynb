1) What is K-Nearest Neighbors (KNN) and how does it work?

K-Nearest Neighbors (KNN) is a simple, non-parametric, and lazy supervised learning algorithm used for classification and regression tasks.

How it works:
Classification: Given a new data point, KNN searches for the 'K' closest data points in the training set based on a distance metric (usually Euclidean distance).

Decision: For classification, the most common class among the K nearest neighbors is assigned to the new data point. For regression, the average of the K nearest neighbors' target values is taken as the prediction.

No Training Phase: KNN does not involve a model-building phase. It stores the training dataset and makes decisions based on proximity at the time of prediction.

Key Points:
K: The number of neighbors to consider.

Distance Metric: Often Euclidean distance, but other metrics can be used.



2) What is the difference between KNN Classification and KNN Regression?

The key difference between KNN Classification and KNN Regression lies in the type of output they predict:

KNN Classification:
Objective: Predict the class or category of a given data point.

How it works:

For a given test point, KNN looks at the 'K' nearest neighbors from the training data.

The class label that is most common among the K neighbors is assigned to the test point.

Output: Categorical value (class label).

Example: Predicting whether an email is spam or not spam.

KNN Regression:
Objective: Predict a continuous value (a real number).

How it works:

For a given test point, KNN looks at the 'K' nearest neighbors from the training data.

The predicted value is typically the mean or average of the target values of the K nearest neighbors.

Output: Continuous value (real number).

Example: Predicting the price of a house based on features like area, number of rooms, etc.

Summary:
KNN Classification: Assigns a class label.

KNN Regression: Predicts a continuous numeric value.

3) What is the role of the distance metric in KNN?

The distance metric in KNN determines how similarity between data points is measured. It helps identify the "nearest" neighbors to a query point.

Euclidean distance: Measures straight-line distance (common for continuous data).

Manhattan distance: Measures sum of absolute differences (useful for grid-like data).

Cosine similarity: Measures angle between vectors (used in text or sparse data).

The choice of metric influences model performance and the shape of decision boundaries.

4) What is the Curse of Dimensionality in KNN?

The Curse of Dimensionality in KNN refers to the phenomenon where the performance of the algorithm deteriorates as the number of features (dimensions) increases.

In high-dimensional spaces:

Distance between points becomes similar: As the number of dimensions increases, data points become sparse, and the distance between neighbors becomes less meaningful.

Increased computational cost: With more features, the algorithm requires more time and memory to compute distances.

Overfitting: High-dimensional data can lead to overfitting because KNN may fit to noise or irrelevant features, resulting in poor generalization.

In essence, as the dimensionality grows, the data becomes less structured, and KNN's ability to distinguish between neighbors weakens.

5)  How can we choose the best value of K in KNN?

To choose the best value of K in KNN, you can use the following methods:

Cross-validation: Split the data into multiple subsets and evaluate the model's performance for different values of K using cross-validation. The K that gives the best average performance is typically chosen.

Error vs K Plot: Plot the error rate (or accuracy) for different values of K. Choose the K with the lowest error (or highest accuracy) on the validation set.

Odd values of K: For classification, use odd values of K to avoid ties when assigning class labels.

Balance Bias-Variance: A small K leads to low bias but high variance (overfitting), while a large K reduces variance but increases bias (underfitting). Select K that balances these.

In short, use cross-validation or error plots to find the K that minimizes error or maximizes accuracy.

6)  What are KD Tree and Ball Tree in KNN?

KD Tree and Ball Tree are data structures used in KNN to efficiently search for the nearest neighbors, especially in high-dimensional spaces.

KD Tree (K-Dimensional Tree):

A binary tree that partitions the data space into two halves based on feature values.

It works well for low to medium-dimensional data.

Splits the data recursively along one feature at each level (e.g., splitting by x-axis, then y-axis, etc.).

Faster than brute-force search, but performance degrades as dimensionality increases (due to the curse of dimensionality).

Ball Tree:

A data structure that partitions the data into balls (clusters) in a multi-dimensional space.

Works well for high-dimensional data compared to KD Tree.

More efficient for finding nearest neighbors when there are many dimensions.

Both structures help speed up the neighbor search by reducing the number of distance computations required.

7) When should you use KD Tree vs. Ball Tree?

Use KD Tree when:

The data has low to medium dimensions (typically fewer than 20-30 features).

The dataset is relatively well-distributed and doesn't have many outliers.

Use Ball Tree when:

The data has high dimensions (more than 30 features).

The data is sparse or has many clusters, as Ball Tree is better suited for high-dimensional spaces.

In general, KD Tree is more efficient for lower-dimensional data, while Ball Tree is preferred for higher-dimensional or sparse data.

8) What are the disadvantages of KNN?

The main disadvantages of KNN are:

Computationally Expensive: As KNN stores all the training data and computes distances during prediction, it becomes slow with large datasets.

Curse of Dimensionality: KNN performance degrades with high-dimensional data, as distances between points become less meaningful.

Sensitive to Noise: KNN can be heavily influenced by noisy or irrelevant features, leading to incorrect predictions.

Memory Intensive: Since it needs to store all training data, KNN can be memory-intensive, especially with large datasets.

Poor Performance on Imbalanced Data: KNN can perform poorly when the classes in the dataset are imbalanced, as it may give more weight to the majority class.

9) How does feature scaling affect KNN?

Feature scaling significantly impacts KNN because the algorithm relies on distance metrics (like Euclidean distance) to determine the similarity between data points.

Without scaling: Features with larger numerical ranges dominate the distance calculation, leading to biased results. For example, a feature with values ranging from 1 to 1000 will have a disproportionate influence on the distance compared to a feature with values between 0 and 1.

With scaling: All features contribute equally to the distance metric, improving KNN's performance. Standard scaling (e.g., Min-Max scaling or Standardization) ensures that each feature has the same weight in the distance calculation.

In short, feature scaling is essential in KNN to ensure that each feature contributes equally to the distance metric and prevents any one feature from dominating.

10) What is PCA (Principal Component Analysis)?

PCA (Principal Component Analysis) is a dimensionality reduction technique used to simplify large datasets while preserving as much variance (information) as possible.

It transforms the original features into a new set of uncorrelated features called principal components.

These components are ordered by importance, meaning the first few capture most of the variation in the data.

PCA helps in reducing noise, improving visualization, and speeding up algorithms like KNN by reducing dimensionality.


11)  How does PCA work?

PCA works by transforming the original data into a new coordinate system where:

It centers the data (subtracts the mean).

It calculates the covariance matrix to understand how features vary together.

It computes eigenvectors and eigenvalues of this matrix to find directions (principal components) of maximum variance.

It projects the data onto the top principal components (those with highest variance).

12) What is the geometric intuition behind PCA?

The geometric intuition behind PCA is that it rotates the coordinate system to align with the directions of maximum variance in the data.

Imagine a cloud of data points in high-dimensional space — PCA finds new axes (principal components) that:

Point in the directions where the data spreads out the most.

Are orthogonal (at right angles) to each other.

Allow projection of the data onto fewer axes with minimal loss of information.

13) What is the difference between Feature Selection and Feature Extraction?

Feature Selection and Feature Extraction are both techniques for dimensionality reduction, but they work differently:

Feature Selection: Chooses a subset of existing features based on importance (e.g., using correlation, mutual information, or model-based methods). No transformation is applied.

Feature Extraction: Transforms the original features into a new set of features (e.g., PCA), often combining information from multiple original features.

14)  What are Eigenvalues and Eigenvectors in PCA?

In PCA:

Eigenvectors represent the directions (principal components) along which the data varies the most.

Eigenvalues indicate the amount of variance captured by each eigenvector.

15) How do you decide the number of components to keep in PCA?

To decide the number of components to keep in PCA, you can:

Use the explained variance ratio: Choose enough components to retain a desired percentage (e.g., 95%) of the total variance.

Plot a scree plot: Look for the "elbow point" where adding more components gives diminishing returns.

In short: Keep the number of components that retain most of the data’s variance (often 95% or more) using explained variance or scree plot.

16) Can PCA be used for classification?

PCA itself is not a classification algorithm, but it can be used before classification to reduce dimensionality, remove noise, and improve performance.

17) What are the limitations of PCA?

The main limitations of PCA are:

Linear only: PCA assumes linear relationships and may not capture complex, non-linear patterns.

Loss of interpretability: The new features (principal components) are combinations of original features and hard to interpret.

Sensitive to scaling: PCA requires features to be standardized; otherwise, results can be misleading.


18) How do KNN and PCA complement each other?

KNN and PCA complement each other well:

PCA reduces dimensionality, removes noise, and helps prevent the curse of dimensionality.

This makes KNN more efficient and accurate, as distance calculations become more meaningful in lower-dimensional space.

19) How does KNN handle missing values in a dataset?

KNN does not handle missing values directly — they must be addressed before applying the algorithm.

Common strategies include:

Imputing missing values using KNN imputation (finding nearest neighbors and averaging their values).

Removing rows/columns with missing data if appropriate.

20) What are the key differences between PCA and Linear Discriminant Analysis (LDA)?

The key differences between PCA and LDA are:

Purpose:

PCA: Aims to maximize variance in the data, focusing on capturing the most significant directions regardless of class labels.

LDA: Aims to maximize class separability, focusing on finding the axes that best separate different classes.

Supervision:

PCA: Unsupervised method, doesn't consider class labels.

LDA: Supervised method, takes class labels into account during transformation.

Output:

PCA: Creates new components that maximize total variance in the data.

LDA: Creates new components that maximize the separation between classes.

Dimensionality:

PCA: Can produce more components than the number of classes.

LDA: The number of components is at most C - 1, where C is the number of classes.

```python
# Train a KNN Classifier on the Iris dataset and print model accuracy?

# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model
knn.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test)

# Calculate and print accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of the KNN Classifier: {accuracy:.4f}')
```

```python
# Train a KNN Classifier with different values of K and visualize decision boundaried?

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Use only the first two features for visualization
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a meshgrid for plotting decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

# Function to plot decision boundaries
def plot_decision_boundary(K):
    # Train KNN classifier with the given value of K
    knn = KNeighborsClassifier(n_neighbors=K)
    knn.fit(X_train, y_train)

    # Predict class for each point in the meshgrid
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the decision boundary
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=ListedColormap(['#FFAAAA', '#AAAAFF', '#AAFFAA']))
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, edgecolors='k', marker='o', cmap=ListedColormap(['red', 'blue', 'green']))
    plt.title(f'KNN Classifier with K = {K}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# Visualize decision boundaries for different values of K
for K in [1, 3, 5, 7]:
    plot_decision_boundary(K)
```

```python
# Apply PCA before training a KNN Classifier and compare accuracy with and without PCA?

# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Apply feature scaling (Standardization) before PCA and KNN
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train KNN model without PCA
knn_unscaled = KNeighborsClassifier(n_neighbors=3)
knn_unscaled.fit(X_train_scaled, y_train)
y_pred_unscaled = knn_unscaled.predict(X_test_scaled)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 components for visualization
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Train KNN model with PCA
knn_pca = KNeighborsClassifier(n_neighbors=3)
knn_pca.fit(X_train_pca, y_train)
y_pred_pca = knn_pca.predict(X_test_pca)
accuracy_pca = accuracy_score(y_test, y_pred_pca)

# Print the accuracies
print(f'Accuracy of KNN Classifier without PCA: {accuracy_unscaled:.4f}')
print(f'Accuracy of KNN Classifier with PCA: {accuracy_pca:.4f}')
```

```python
#  Perform Hyperparameter Tuning on a KNN Classifier using GridSearchCV?
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the KNN classifier
knn = KNeighborsClassifier()

# Define the hyperparameters grid to search over
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],  # Different number of neighbors
    'metric': ['euclidean', 'manhattan'],  # Different distance metrics
    'weights': ['uniform', 'distance']  # Different weight strategies
}

# Apply GridSearchCV
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Get the best hyperparameters and the best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Evaluate the best model on the test set
best_knn = grid_search.best_estimator_
y_pred = best_knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

# Print the best hyperparameters and accuracy
print(f'Best Hyperparameters: {best_params}')
print(f'Best Cross-Validation Score: {best_score:.4f}')
print(f'Accuracy on Test Set: {accuracy:.4f}')
```

```python
#Train a PCA model and visualize the cumulative explained variance?

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train a PCA model
pca = PCA()
pca.fit(X_scaled)

# Get the cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot the cumulative explained variance
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
# Train a KNN Classifier using different values of the weights parameter (uniform vs. distance) and compare accuracy

# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize KNN classifiers with different weight strategies
knn_uniform = KNeighborsClassifier(n_neighbors=3, weights='uniform')
knn_distance = KNeighborsClassifier(n_neighbors=3, weights='distance')

# Train the models
knn_uniform.fit(X_train, y_train)
knn_distance.fit(X_train, y_train)

# Make predictions
y_pred_uniform = knn_uniform.predict(X_test)
y_pred_distance = knn_distance.predict(X_test)

# Calculate the accuracy for both models
accuracy_uniform = accuracy_score(y_test, y_pred_uniform)
accuracy_distance = accuracy_score(y_test, y_pred_distance)

# Print the results
print(f'Accuracy with uniform weights: {accuracy_uniform:.4f}')
print(f'Accuracy with distance weights: {accuracy_distance:.4f}')
```

```python
#4 Train a KNN Regressor and analyze the effect of different K values on performance?

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Create a synthetic regression dataset
X, y = make_regression(n_samples=100, n_features=1, noise=0.1, random_state=42)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# List to store Mean Squared Errors for different K values
k_values = range(1, 21)  # Test K values from 1 to 20
mse_values = []

# Train KNN Regressor with different K values and calculate MSE
for k in k_values:
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)

# Plot MSE against K values
plt.figure(figsize=(8, 6))
plt.plot(k_values, mse_values, marker='o', linestyle='-', color='b')
plt.title('Effect of K value on KNN Regressor Performance')
plt.xlabel('K value')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid(True)
plt.show()

# Print the K with the lowest MSE
best_k = k_values[np.argmin(mse_values)]
print(f'Best K value with the lowest MSE: {best_k}')
```

```python
# Train a KNN Classifier using the KD Tree and Ball Tree algorithms and compare performance?

# Import necessary libraries
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the KNN classifier using KD Tree algorithm
knn_kd_tree = KNeighborsClassifier(n_neighbors=3, algorithm='kd_tree')
knn_kd_tree.fit(X_train, y_train)
y_pred_kd_tree = knn_kd_tree.predict(X_test)
accuracy_kd_tree = accuracy_score(y_test, y_pred_kd_tree)

# Train the KNN classifier using Ball Tree algorithm
knn_ball_tree = KNeighborsClassifier(n_neighbors=3, algorithm='ball_tree')
knn_ball_tree.fit(X_train, y_train)
y_pred_ball_tree = knn_ball_tree.predict(X_test)
accuracy_ball_tree = accuracy_score(y_test, y_pred_ball_tree)

# Print the accuracy for both algorithms
print(f'Accuracy with KD Tree algorithm: {accuracy_kd_tree:.4f}')
print(f'Accuracy with Ball Tree algorithm: {accuracy_ball_tree:.4f}')
```

```python
#  Train a PCA model and analyze the effect of different numbers of components on accuracy?

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)

# List to store accuracy scores
accuracy_scores = []

# Iterate over different numbers of PCA components (from 1 to 4)
for n_components in range(1, X.shape[1] + 1):
    # Apply PCA with n_components
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    # Train the KNN model
    knn.fit(X_train_pca, y_train)

    # Make predictions on the test set
    y_pred = knn.predict(X_test_pca)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Plot accuracy vs. number of components
plt.figure(figsize=(8, 6))
plt.plot(range(1, X.shape[1] + 1), accuracy_scores, marker='o', linestyle='-', color='b')
plt.title('Effect of Number of PCA Components on KNN Classifier Accuracy')
plt.xlabel('Number of PCA Components')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
# Train a KNN Classifier with different leaf_size values and compare accuracy?

# Import necessary libraries
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# List to store accuracy scores for different leaf_size values
leaf_sizes = [10, 20, 30, 40, 50]
accuracy_scores = []

# Iterate over different leaf_size values
for leaf_size in leaf_sizes:
    # Initialize the KNN classifier with the current leaf_size
    knn = KNeighborsClassifier(n_neighbors=3, algorithm='ball_tree', leaf_size=leaf_size)

    # Train the model
    knn.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = knn.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Plot the accuracy comparison
plt.figure(figsize=(8, 6))
plt.plot(leaf_sizes, accuracy_scores, marker='o', linestyle='-', color='b')
plt.title('Effect of leaf_size on KNN Classifier Accuracy')
plt.xlabel('leaf_size')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()

# Print accuracy scores
for leaf_size, accuracy in zip(leaf_sizes, accuracy_scores):
    print(f'Accuracy with leaf_size={leaf_size}: {accuracy:.4f}')
```

```python
#Train a KNN Regressor and analyze the effect of different distance metrics on prediction error?
# Import necessary libraries
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load the Diabetes dataset (regression problem)
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# List of distance metrics to compare
metrics = ['euclidean', 'manhattan']
errors = []

# Train KNN Regressor with different distance metrics and calculate MSE
for metric in metrics:
    # Initialize the KNN regressor with the current distance metric
    knn = KNeighborsRegressor(n_neighbors=5, metric=metric)

    # Train the model
    knn.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = knn.predict(X_test)

    # Calculate the Mean Squared Error
    mse = mean_squared_error(y_test, y_pred)
    errors.append(mse)

# Plot the comparison of MSE for different metrics
plt.figure(figsize=(8, 6))
plt.bar(metrics, errors, color='skyblue')
plt.title('Effect of Distance Metrics on KNN Regressor MSE')
plt.xlabel('Distance Metric')
plt.ylabel('Mean Squared Error')
plt.show()

# Print the MSE for each metric
for metric, error in zip(metrics, errors):
    print(f'Mean Squared Error with {metric} distance: {error:.4f}')
```

```python
# Train a KNN Classifier and evaluate using ROC-AUC score?
# Import necessary libraries
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Binarize the output labels for multi-class classification
y_bin = label_binarize(y, classes=[0, 1, 2])

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_bin, test_size=0.3, random_state=42)

# Initialize the KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model
knn.fit(X_train, y_train)

# Make predictions on the test set (probabilities for ROC-AUC calculation)
y_prob = knn.predict_proba(X_test)

# Compute the ROC-AUC score for each class
roc_auc = {}
for i in range(y_bin.shape[1]):
    roc_auc[i] = roc_auc_score(y_test[:, i], y_prob[i][:, 1])

# Print the ROC-AUC scores for each class
for i in range(len(roc_auc)):
    print(f"ROC-AUC Score for Class {i}: {roc_auc[i]:.4f}")

# Plot ROC Curve for each class
fpr, tpr, _ = roc_curve(y_test[:, 0], y_prob[0][:, 1])
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'Class 0 (AUC = {roc_auc[0]:.2f})')
fpr, tpr, _ = roc_curve(y_test[:, 1], y_prob[1][:, 1])
plt.plot(fpr, tpr, color='red', label=f'Class 1 (AUC = {roc_auc[1]:.2f})')
fpr, tpr, _ = roc_curve(y_test[:, 2], y_prob[2][:, 1])
plt.plot(fpr, tpr, color='green', label=f'Class 2 (AUC = {roc_auc[2]:.2f})')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for KNN Classifier')
plt.legend(loc='lower right')
plt.show()
```

```python
# Train a KNN Classifier and perform feature selection before training?
# Import necessary libraries
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Perform feature selection using SelectKBest with chi-squared test
selector = SelectKBest(chi2, k=2)  # Select top 2 features
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

# Initialize the KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)

# Train the KNN classifier on the selected features
knn.fit(X_train_selected, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test_selected)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy of the model
print(f'Accuracy of KNN Classifier after Feature Selection: {accuracy:.4f}')

# Print the selected features
print(f'Selected Features: {selector.get_support()}')
```

```python
# Train a PCA model and visualize the data reconstruction error after reducing dimensions?
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Standardize the features (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize PCA and fit to the data
pca = PCA(n_components=2)  # Reducing to 2 components for visualization
X_pca = pca.fit_transform(X_scaled)

# Reconstruct the data from the reduced components
X_reconstructed = pca.inverse_transform(X_pca)

# Calculate the reconstruction error (mean squared error)
reconstruction_error = mean_squared_error(X_scaled, X_reconstructed)

# Print the reconstruction error
print(f'Reconstruction Error (Mean Squared Error): {reconstruction_error:.4f}')

# Visualize the original data and the reconstructed data (first two components)
plt.figure(figsize=(8, 6))

# Plot the original data in 2D (first two features)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], label='Original Data', alpha=0.6, color='b')

# Plot the reconstructed data in 2D (first two components)
plt.scatter(X_reconstructed[:, 0], X_reconstructed[:, 1], label='Reconstructed Data', alpha=0.6, color='r')

plt.title('Original Data vs Reconstructed Data (PCA)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
#Train a KNN Classifier and visualize the decision boundary?

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Use only the first two features for visualization (sepal length and sepal width)
y = iris.target

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Initialize the KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)

# Train the KNN classifier
knn.fit(X_train, y_train)

# Create a mesh grid to plot the decision boundary
h = .02  # Step size in the mesh
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predict on the mesh grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, edgecolors='k', marker='o', cmap=plt.cm.RdYlBu)
plt.title("KNN Classifier Decision Boundary")
plt.xlabel('Feature 1 (Sepal length)')
plt.ylabel('Feature 2 (Sepal width)')
plt.colorbar()
plt.show()
```

```python
# Train a PCA model and analyze the effect of different numbers of components on data variance?

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardize the features (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize PCA with a maximum number of components equal to the number of features
pca = PCA()

# Fit PCA on the scaled data
pca.fit(X_scaled)

# Calculate the explained variance ratio for each component
explained_variance_ratio = pca.explained_variance_ratio_

# Plot the explained variance ratio
plt.figure(figsize=(8, 6))

# Plot individual explained variance ratios
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.6, color='b', label='Individual Explained Variance')

# Plot cumulative explained variance
cumulative_variance = np.cumsum(explained_variance_ratio)
plt.plot(range(1, len(explained_variance_ratio) + 1), cumulative_variance, color='r', marker='o', label='Cumulative Explained Variance')

plt.title('PCA: Explained Variance vs Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
